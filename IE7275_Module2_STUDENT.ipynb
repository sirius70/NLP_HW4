#### **Exercise 2: Basic Python Practice — Variables, Loops, Functions, and Data Structures**

In this exercise, you will practice using basic Python syntax by creating a list of five integers and computing their sum with a for loop, writing a function named `find_max` that returns the maximum value from a list of numbers, and creating a dictionary with three fruits and their quantities, then printing each key-value pair using a for loop.


```python
# Create a list of five integers
numbers = [10, 20, 30, 40, 50]
total_sum = 0

# --- Added the code here ---
# Compute their sum using a for loop
# ---------------------------
for num in numbers:
    total_sum += num

print("Sum of numbers:", total_sum)
```

```python
# Write a function that returns the maximum value in a list
def find_max(num_list):
    # --- Added the code here ---
    max_value = num_list[0]
    # ---------------------------
    for num in num_list:
        if num > max_value:
            max_value = num
    return max_value

# Test the function
print("Maximum value:", find_max(numbers))
```

#### **Exercise 10: Measuring Skewness and Applying Z-Scores for Outlier Detection**

In this exercise, you will first generate a new DataFrame with Score values that include outliers. Then, you will compute and interpret skewness, and apply z-score based outlier detection to identify extreme values. This will help you clearly see how z-scores work for identifying outliers.


```python
# Import necessary libraries
import pandas as pd
import numpy as np
from scipy.stats import zscore

# Create a new DataFrame with outliers
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva', 'Frank', 'Grace', 'Hannah'],
    'Age': [25, 30, 22, 35, 28, 27, 23, 29],
    'Score': [88, 92, 79, 45, 90, 300, 5, 83]
}

df = pd.DataFrame(data)
print(df)
```

```python
# Compute and print skewness of 'Score'

# --- Added the code here ---
skewness_score = df['Score'].skew()
# ---------------------------
print("Skewness of Score:", skewness_score)
```

---
#### **Question:**

In the script, we calculated the skewness of the 'Score' column to understand the shape of the data distribution.

**Why is it helpful to calculate skewness when analyzing a dataset, and how can skewness affect your interpretation or modeling of the data?**  
> Explain your answer clearly in your own words.

---
Skewness is useful because it tells us if the data is lopsided or balanced. If the data is skewed, it means some very high or very low values are stretching the distribution in one direction. This matters because skewed data can make averages misleading and affect how well models work. By knowing the skew, we can decide whether to adjust the data, handle outliers, or pick models that deal better with extreme values, so our analysis and predictions are more reliable.


```python
# --- Added the code here ---
df['Score_zscore'] = zscore(df['Score']) # Compute z-scores for 'Score'
# ---------------------------

# --- Added the code here ---
outliers = df[np.abs(df['Score_zscore']) > 2] # Identify potential outliers with lower threshold (|z| > 2)
# ---------------------------

print("Potential outliers based on z-scores:\n", outliers)
```

#### **Exercise 15: Handling Severe Class Imbalance with Resampling and Using Appropriate Metrics**

In this exercise, you will generate a highly imbalanced classification dataset, apply a simple resampling strategy (random oversampling of the minority class), train a simple classification model (logistic regression), and evaluate it using precision, recall, and F1-score — going beyond accuracy to better assess model performance on the minority class.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report

# Step 1: Create a highly imbalanced DataFrame
data = {
    'Feature1': np.random.randn(100),
    'Feature2': np.random.randn(100),
    'target': ['Majority'] * 95 + ['Minority'] * 5
}

df = pd.DataFrame(data)
print(df)
```

```python
# Step 2: Visualize and print initial class distribution
print("Original Class Distribution:\n", df['target'].value_counts())
print("\nOriginal Class Percentages (%):\n", df['target'].value_counts(normalize=True) * 100)

df['target'].value_counts().plot(kind='bar', color='orange', edgecolor='black')
plt.title('Original Class Distribution')
plt.xlabel('Class')
plt.ylabel('Frequency')
plt.grid(axis='y')
plt.show()
```

```python
from sklearn.utils import resample

# Step 3: Split features and target (before resampling)
X = df[['Feature1', 'Feature2']]
y = df['target']

# Step 4: Perform random oversampling of the minority class
df_minority = df[df['target'] == 'Minority']
df_majority = df[df['target'] == 'Majority']

# --- Added the code here ---
df_minority_oversampled = resample(df_minority, n_samples=95, replace=True, random_state=42)
# ---------------------------

# Combine majority and oversampled minority

# --- Added the code here ---
df_resampled = pd.concat([df_majority, df_minority_oversampled])
# ---------------------------

# Shuffle the DataFrame
df_resampled = df_resampled.sample(frac=1, random_state=42).reset_index(drop=True)

# Step 5: Visualize and print class distribution after resampling
print("\nResampled Class Distribution:\n", df_resampled['target'].value_counts())
print("\nResampled Class Percentages (%):\n", df_resampled['target'].value_counts(normalize=True) * 100)

df_resampled['target'].value_counts().plot(kind='bar', color='green', edgecolor='black')
plt.title('Resampled Class Distribution')
plt.xlabel('Class')
plt.ylabel('Frequency')
plt.grid(axis='y')
plt.show()
```

---
#### **Question:**

In the script, we performed random oversampling to balance the class distribution and then shuffled the dataset.

**Why is balancing the classes through techniques like oversampling important when training a machine learning model on imbalanced data?**  
> Explain your answer clearly in your own words.

---
Balancing classes is important because if one class has way more examples than the others, the model will mostly learn to predict the majority class and ignore the minority class. Oversampling the minority class gives the model enough examples to learn from, so it can recognize all classes fairly. This helps the model make better predictions and avoids bias toward the bigger class. Shuffling the data afterward ensures the model sees a good mix of classes during training.


```python
from sklearn.linear_model import LogisticRegression

# Step 6: Split resampled data into features and target
X_resampled = df_resampled[['Feature1', 'Feature2']]
y_resampled = df_resampled['target']

# Step 7: Train-test split
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.3, random_state=42)

# Step 8: Train a simple Logistic Regression model
print("\nTraining Logistic Regression model...")

# --- Added the code here ---
model = LogisticRegression(solver='liblinear')
# ---------------------------

model.fit(X_train, y_train)

# Step 9: Make predictions and print evaluation metrics
y_pred = model.predict(X_test)

print("\nClassification Report (Precision, Recall, F1-score):\n")
print(classification_report(y_test, y_pred))
```
